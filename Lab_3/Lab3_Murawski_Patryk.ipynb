```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cars = pd.read_csv('samochody1tys.csv')
```

```python
#Wykres funkcji linowej 
x = [1,4,8]
y = [2,5,8]
plt.plot(x,y,'r',label='liniowa')
plt.title('Funkcja liniowa')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
```

```python
#wykres funkcji kwadratowej
x2 = np.linspace(-10,10,100)
plt.plot(x2, x2**2, label='kwadratowa')
plt.title("Funkcji kwadratowa")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
```

```python
#wykres funkcji  trzeciego rzdu
plt.plot(x2, x2**3, 'p', label='trzeciego rzdu')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Funkcja trzeciego rzdu")
plt.legend()
plt.grid(True)
```

```python
#wykres funkcji sinus
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y,'--')
plt.title('Funkcja sinus')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['sin(x)'])
plt.grid(True)
```

```python
#Wykres supkowy iloci samochodw danej marki 
marka = cars['marka'].value_counts().sort_values()
marka.plot.bar(figsize=(25,10))
plt.title('Liczebno pojazdw kadej marki')
plt.xlabel('Marka pojazdu')
plt.ylabel('Ilo pojazdw')
plt.show()
```

```python
paliwo = pd.Index(cars['rodzaj_silnika']).value_counts()
paliwo
```

```python
#Wykres koowy iloci wystpie kadego rodzaju silnika
paliwo = pd.Index(cars['rodzaj_silnika']).value_counts()
paliwo.plot.pie(autopct='%1.1f%%',   figsize=(25,10))
plt.title('Rodzaje silnikw w wymienionych pojazdach')
plt.show()
```

```python
#Histogram roku produkcji samochodw(wykres ukazujcy czstotliow ukazywania si danej wartoci)
woj = cars['wojewodztwo']
histogram = woj.hist( figsize=(30,10))
histogram.set_title('Histogram iloci aut w wojewdztwach')
histogram.set_xlabel('Wojewdztwo')
histogram.set_ylabel('Ilo')
plt.show()
```

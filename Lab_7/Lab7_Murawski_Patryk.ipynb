```python
import numpy as np
import pandas as pd
```

```python
def activate(x):
    if x>0:
        return 1
    return 0
```

```python
print("Bramka AND")
bias = 1 * -1.5
bias_list = [-1.5,-1.5,-1.5,-1.5]
x1 = [0,0,1,1]
x2 = [0,1,0,1]
x1_weight = 1
x2_weight = 1
logic_values = [x1*x2 for x1,x2 in zip(x1,x2)]
and_sum = [x1+x2+b for x1,x2,b in zip(x1,x2,bias_list)]
output = [activate(x) for x in and_sum]
result = pd.DataFrame({'X1':x1,
                      'X2':x2,
                      'X1&X2':logic_values,
                      'X1 weight':x1_weight,
                      'X2 weight':x2_weight,
                      'bias': bias,
                      'and_sum':and_sum,
                      'output(a)': output})
result
```

```python
print("Bramka OR")
def OR(a, b):
    return 1 if a == 1 else 1 if b == 1 else 0
    
bias = 1 * -0.5
bias_list = [-0.5,-0.5,-0.5,-0.5]
x1 = [0,0,1,1]
x2 = [0,1,0,1]
x1_weight = 1
x2_weight = 1
logic_values = [OR(x1,x2) for x1,x2 in zip(x1,x2)]
or_sum = [x1+x2+b for x1,x2,b in zip(x1,x2,bias_list)]
output = [activate(x) for x in or_sum]
result = pd.DataFrame({'X1':x1,
                      'X2':x2,
                      'X1||X2':logic_values,
                      'X1 weight':x1_weight,
                      'X2 weight':x2_weight,
                      'bias': bias,
                      'or_sum':or_sum,
                      'output(a)': output})
result
```

```python
print("Bramka NOT")
def NOT(x):
    return 1 if x == 0 else 0 if x == 1 else None

bias = 1 * 1
bias_list = [1,1]
x1 = [0,1]
x1_weight = [-2,-2]
logic_values = [NOT(x) for x in x1]
not_sum = [b + x * y for b,x,y in zip(bias_list,x1,x1_weight)]
output = [activate(x) for x in not_sum]
result = pd.DataFrame({'X1':x1,
                      '!X1':logic_values,
                      'X1 weight':x1_weight,
                      'bias': bias,
                      'not_sum':not_sum,
                      'output(a)': output})
result
```

```python
print("Bramka XNOR")
def XNOR(a,b):
    return 1 if a == b else 0

bias = 1 * 0.5 #0.5
bias_list = [0.5,0.5,0.5,0.5]
x1 = [0,0,1,1]
x2 = [0,1,0,1]
x1_weight = -1
x2_weight = -1
x1_not = [NOT(x) for x in x1]
x2_not = [NOT(x) for x in x2]
logic_values = [x1*x2 for x1,x2 in zip(x1_not,x2_not)]
activation_sum = [b-x1-x2 for b,x1,x2 in zip(bias_list,x1,x2)]
a1 = [activate(x) for x in activation_sum]
a2 = [x1*x2 for x1,x2 in zip(x1,x2)]
a3 = [XNOR(x1,x2) for x1,x2 in zip(x1,x2)]
result = pd.DataFrame({'X1':x1,
                        'X2':x2,
                        'logic_values':logic_values,
                        'X1 weight':x1_weight,
                        'X2 weight':x1_weight,
                        'bias': bias,
                        'activation_sum':activation_sum,
                        'a1': a1,
                        'a2':a2,
                        'a3':a3})
result
```
